```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
```

```python
stock_prices = pd.read_csv("regression_predicted.csv")
stock_prices
```

```python
stock_prices["Date"] = pd.to_datetime(stock_prices["Date"])
```

```python
import pandas as pd

all_stocks_df = pd.DataFrame()

for code in stock_prices["SecuritiesCode"].unique():
    tmpdf = stock_prices[stock_prices["SecuritiesCode"] == code].reset_index(drop=True)

    all_stocks_df = pd.concat([all_stocks_df, tmpdf], ignore_index=True)

```

```python
all_stocks_df["Close_shift1"] = all_stocks_df["prediction"].shift(-1)
all_stocks_df["Close_shift2"] = all_stocks_df["prediction"].shift(-2)

all_stocks_df["rate"] = (all_stocks_df["Close_shift2"] - all_stocks_df["Close_shift1"]) / all_stocks_df["Close_shift1"]
all_stocks_df
```

```python
all_stocks_df["rank"] = all_stocks_df["rate"].rank(ascending=False,method="first") -1 
all_stocks_df = all_stocks_df.sort_values("rank").reset_index(drop=True)
```
